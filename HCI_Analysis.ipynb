```
n_people, n_tasks = click_c.shape
```

```
for i in range(n_tasks):
    print("Task %d" % (i+1))
    print("Average time taken with Collab:", np.mean(time_c[:, i]))
    print("Average time taken with OWCA:", np.mean(time_d[:, i]))
    print()
```

```
# Assuming 2 courses, 14 weeks of classes, 2 classes per week
# How much time do you save using our dashboard?
difference = (np.mean(time_c[:, 0]) - np.mean(time_d[:, 0])) * 2 * 14 * 2
collab_total = np.mean(time_c[:, 0]) * 2 * 14 * 2
ours_total   = np.mean(time_d[:, 0]) * 2 * 14 * 2
print("Time on Collab: %d minutes, %d seconds" % (collab_total/60, int(collab_total) % 60))
print("Time on our Design: %d minutes, %d seconds" % (ours_total/60, int(ours_total) % 60))
print("Time saved: %d minutes, %d seconds" % (difference/60, int(difference) % 60))
```

```
for i in range(n_tasks):
    print("Task %d" % (i+1))
    print("Average clicks with Collab:", np.mean(click_c[:, i]))
    print("Average clicks with OWCA:", np.mean(click_d[:, i]))
    print()
```

```
# Check if variances are same for tasks
for i in range(n_tasks):
  stat, p = stats.levene(time_c[:, i], time_d[:, i])
  print(stat, p)
print("Variances are very much different!")
```

```
# T-test per Task (time)
for i in range(n_tasks):
  stat, pvalue = stats.ttest_rel(time_c[:, i], time_d[:, i])
  # Since this is a 2-tail test, scale P down to 1-sided
  # Reference: https://stackoverflow.com/questions/15984221/how-to-perform-two-sample-one-tailed-t-test-with-numpy-scipy
  print("Task %d: T-test value: %.3f with p-value=%.3f" % (i+1, stat, pvalue/2))

print()
print("T-values are > 0 for all of the tasks")
print("p-values are < 0.01 for 4/5 tasks, implying 99% confidence in t-test")
print("p-values is < 0.05 for 1/5 task, implying 95% confidence in t-test")
```

```
# T-test per Task (clicks)
for i in range(n_tasks):
  stat, pvalue = stats.ttest_rel(click_c[:, i], click_d[:, i])
  # Since this is a 2-tail test, scale P down to 1-sided
  # Reference: https://stackoverflow.com/questions/15984221/how-to-perform-two-sample-one-tailed-t-test-with-numpy-scipy
  print("Task %d: T-test value: %.3f with p-value=%.3f" % (i+1, stat, pvalue/2))

print()
print("T-values are > 0 for all of the tasks")
print("p-values are < 0.01 for 5/5 tasks, implying 99% confidence in t-test")
```

```
# Wilcoxon's test per Task (time)
for i in range(n_tasks):
  stat, pvalue = stats.wilcoxon(time_c[:, i], time_d[:, i], alternative='greater')
  print("Task %d: T-test value: %.3f with p-value=%.3f" % (i+1, stat, pvalue))
```
